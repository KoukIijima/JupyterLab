```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
digits_data = datasets.load_digits()
n = len(digits_data.images)
```

```python
print(n)
```

```python
digits_data['images'][3]
```

```python
image_data = digits_data.images.reshape((n, -1))
```

```python
image_data.shape # それぞれの画像は８*８の配列として表す。
```

```python
image_data[0]
```

```python
labels = digits_data.target
```

```python
pca_transformer = PCA(n_components=0.8)
pca_images = pca_transformer.fit_transform(image_data)
```

```python
pca_images[0]
```

```python
pca_transformer.explained_variance_ratio_
```

```python
sum([0.14890594, 0.13618771, 0.11794594])
```

```python
# 結果を可視化する
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(100):
    ax.scatter(pca_images[i, 0], pca_images[i, 1], pca_images[i, 2], marker=r'${}$'.format(labels[i]), s=64)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(100):
    ax.scatter(pca_images[i, 0], pca_images[i, 1], marker=r'${}$'.format(labels[i]), s=64)
```
